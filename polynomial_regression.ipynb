# Polynomial Regression

It allows us to fit curved line to our data, it is written in the form of

## y = w1*x1 + w2*x2 + w3*x3 + .....+b

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
x = np.arange(0, 30)
y = [3, 4, 5, 7, 10, 8, 9, 10, 10, 23, 27, 44, 50, 63, 67, 60, 62, 70, 75, 88, 81, 87, 95, 100, 108, 135, 151, 160, 169, 179]
```

```python
#visualizing the data
plt.scatter(x, y)
plt.show()
```

```python
poly = PolynomialFeatures(degree=2,include_bias=False) 
#include bias is set to false as linear regression deals with it.

#defining the features
poly_features = poly.fit_transform(x.reshape(-1,1))

#polynomial regression is a linear model, so we import linear regression as well.
poly_reg_model = LinearRegression()
poly_reg_model.fit(poly_features, y)
y_predict = poly_reg_model.predict(poly_features)
```

```python
print("The predicted value is: ", y_predict[:4])
print("The target value is: ", y[:4])
```

```python

plt.title("Polynomial Regression", size=20)
plt.scatter(x, y)
plt.plot(x, y_predict, c="red")
plt.show()
```

```python
#viewing the error
print(r2_score(y, y_predict)) 

#Check the r2 score with different degrees of polynomial to find out how well it fits the data.
```
