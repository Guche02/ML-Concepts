# Linear Regression

It maps a straight line through the data points. It is used to find a possible value from a infinite set of possible values.
The learning algorithm used in uni-variate linear regression is:
## f(x) = w*x + b
Here,
w and b are chosen such that the straight line fits most perfectly through the data points.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# importing the dataset
diabetes_df = pd.read_csv("D:\datasets\diabetes_clean.csv")
print("\n The number of rows and columns in the data: ", diabetes_df.shape)
print(" ")
print(diabetes_df.head())
```

```python
#defining target and feature variables:
X_bmi = diabetes_df["bmi"].values  #.values converts it into a numpy array.
# print(X_bmi)

#reshaping the data to 2d array:
X_bmi = X_bmi.reshape(-1,1)
print(X_bmi.shape)

y = diabetes_df["glucose"].values

# spliting the data into training set and test set:
X_train, X_test, y_train, y_test = train_test_split(X_bmi, y, test_size=0.3, random_state = 42)
# here random state is used to reproduce the same split every time the code is run.

# using linear regression model
reg = LinearRegression()
reg.fit(X_train, y_train)        #training the model.
y_pred = reg.predict(X_test)    #the model is predicting the bmi of the test set
plt.scatter(X_bmi, y, color="Red")
plt.ylabel("Blood Glucose (mg/dl)")
plt.xlabel("Body Mass Index")
plt.plot(X_test, y_pred, color="black")
plt.show()
```

# Error of the model
The R squared is used to analyze the variance in the target values, it ranges from 0 to 1. And, 1 indicates perfect fit.

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```
